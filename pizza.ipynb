```python
'''
    Pizza purchase aplication.
'''

#!/usr/bin/env python
__version__ = '0.1'
__author__ = "NAKAMA"

# Packs
import csv
import datetime
import os
from pathlib import Path
import pickle
import json
size_price_dct = {}
Order_db = "Orders_Database.csv"
```

```python
class Pizza():
    """ Pizza class """
    def __init__(self,Size,Topping=[]):
        self.size = Size
        self.topping = Topping
        self.price = 0
        self.description = """ Pizza is a dish of Italian origin consisting
                                of a usually round, flat base of leavened wheat-based
                                dough topped with tomatoes, cheese, and often various
                                other ingredients (such as various types of sausage,
                                anchovies, mushrooms, onions, olives, vegetables, meat,
                                ham, etc.), which is then baked at a high temperature,
                                traditionally in a wood-fired oven. """

    def get_description(self):
        """ Description of the Pizza class """
        return self.description

    def get_cost(self):
        """ Price of the Pizza class """
        self.price = size_price_dct["size"][self.size]        
        for tpl in self.topping:            
            self.price = self.price + size_price_dct["topping"][tpl]                   
        return self.price

class TurkPizza(Pizza):
    """ Classic Pizza subclass """
    def __init__(self,Size,Topping=["Tomato", "Onions", "Garlic", "Meat"]):
        super(TurkPizza, self).__init__(Size)
        self.topping = Topping
        self.description = """ Turkey has its own traditional variant, the so-called Lahmacun.
                               It is also called Turkish Pizza and is also known as Lahmacun in
                               other countries of the world. A Lahmacun is usually topped with tomato,
                               lamb or ground beef, onions and garlic and seasoned with sumac, cumin,
                               pepper and salt. It is served with a lemon split, spicy Pul Biber
                               (Turkish chili pepper powder) and parsley. Mostly a lahmacun is
                               eaten rolled-up. """    
        
class Margherita(Pizza):
    """ Margherita Pizza subclass """
    def __init__(self,Size,Topping=["Tomato", "GoatCheese", "Basil"]):
        super(Margherita, self).__init__(Size)
        self.topping = Topping
        self.description = """ Pizza Margherita (more commonly known in English as Margherita pizza)
                               is a typical Neapolitan pizza, made with San Marzano tomatoes, mozzarella
                               cheese, fresh basil, salt, and extra-virgin olive oil. Garlic as a topping
                                is omitted. """

class Marinara(Pizza):
    """ Marinara Pizza subclass """
    def __init__(self,Size,Topping=["GoatCheese", "Garlic"]):
        super(Marinara, self).__init__(Size)
        self.topping = Topping
        self.description = """ The Marinara is the purist form of pizza. It is vegan and consists
                                only of dough, San Marzano tomato sauce, garlic and olive oil. """

class PlainPizza(Pizza):
    """ PlainPizza Pizza subclass """
    def __init__(self,Size,Topping=[]):
        super(PlainPizza, self).__init__(Size)
        self.topping = Topping
        self.description = """ A plain pizza would denote no toppings at all and would imply pizza bianca,
                                an unadorned flatbread simply seasoned with salt and olive oil. """
```

```python
class Decorator():
    """ Toppings """
    def __init__(self,Size,Topping=[]):
        self.size = Size
        self.topping = Topping
        self.price = 0
        
    def get_description(self):
        """ Description of Toppings """
        return self.component.get_description() + \
                                       ' ' + Pizza.get_description(self)

    def get_cost(self):
        """ Price of Toppings """                     
        return self.component.get_cost() + \
                                       Pizza.get_cost(self)
```

```python
def get_menu():
    with open(os.path.join(Path.cwd(), "Menu.txt"), mode="r") as file:
        for item in file:
            print(item)
```

```python
def loadData_pickle():
    with open(os.path.join(Path.cwd(), "price.pkl"), mode="rb") as file:
        size_price_dct = pickle.load(file)
        return size_price_dct         

def saveData_pickle():
    size_price_dct = {'small': 90, 'medium': 120, 'large': 150, 'xlarge': 180}
    with open(os.path.join(Path.cwd(), "price.pkl"), mode="wb") as file:        
        pickle.dump(size_price_dct, file) 
```

```python
def loadData_json():
    with open(os.path.join(Path.cwd(), "price.json"), "r") as file:
        size_price_dct = json.load(file)
        return size_price_dct         
        
def saveData_json():
    size_price_dct = {"size" : {"small": 90, "medium": 120, "large": 150, "xlarge": 180} , "topping" : {"small": 90, "medium": 120, "large": 150, "xlarge": 180}}
    with open(os.path.join(Path.cwd(), "price.json"), "w") as file:        
        json.dump(size_price_dct, file, indent = 4, sort_keys = True)
```

```python
def save_order(db,Order,Order_name,Order_ID,Order_cc,Order_ccp,Order_time): #https://www.geeksforgeeks.org/working-csv-files-python/
    # field names
    fields = ['Pizza Size', 'Pizza Type', "Topping", 'Total', 'Name', "ID", 'credit card number', 'credit card password', 'Order Time' ]
    
    with open(db, 'a') as csvfile:
        # creating a csv reader object
        csvreader = csv.reader(csvfile)
        # creating a csv writer object
        csvwriter = csv.writer(csvfile)    
        # data rows of csv file
        rows = [[Order.size, Order.__class__.__name__, Order.topping, Order.get_cost(), Order_name, Order_ID, Order_cc, Order_ccp,Order_time ]]
        # writing the fields
        #csvwriter.writerow(fields)
        # writing the data rows
        csvwriter.writerows(rows)
```

```python
def read_order(db):
    #initializing the titles and rows list
    fields = []
    rows = []
    
    # reading csv file
    with open(db, 'r') as csvfile:
        # creating a csv reader object
        csvreader = csv.reader(csvfile)
        
        # extracting field names through first row
        fields = next(csvreader)
    
        # extracting each data row one by one
        for row in csvreader:
            rows.append(row)
    
        # get total number of rows
        print("Total no. of rows: %d"%(csvreader.line_num))
    
    # printing the field names
    print('Field names are:' + ', '.join(field for field in fields))    
    
    for row in rows:
        # parsing each column of a row
        for col in row:
            print("%10s"%col,end=" "),
        print('\n')

```

```python
def main():
    global size_price_dct
    size_price_dct = loadData_json()
    pizza_topping = []

    while True:
        size_dct = {"1": "small", "2": "medium", "3": "large", "4": "xlarge"}
        print("Pizza Sizes : \n")
        print("1: 'small', 2: 'medium', 3: 'large', 4: 'xlarge' : \n")              
        pizza_size = input ("Select your preferred pizza size ? \n").strip()

        
        if pizza_size in size_dct:
            pizza_size = size_dct[pizza_size]            
            break       
        else:
            print("Sorry, please choose a valid number \n")

    while True:
        get_menu()
        type_dct = {"1": "Marinara", "2": "Margherita", "3": "TurkPizza", "4": "PlainPizza"}
        pizza_type = input ("Select your preferred pizza type ? \n").strip()
        if pizza_type in type_dct:
            pizza_type = type_dct[pizza_type]            
            break       
        else:
            print("Sorry, please choose a valid number \n")
                   
    while True:               
        pt_dct = {"end": "end selection","11": "Olives", "12": "Muhrooms", "13": "GoatCheese", "14": "Meat",\
                   "15": "Onions", "16": "Corn", "17": "Garlic" , "18": "Basil" , "19": "Tomato"}                            
        pt = input ("Select your preferred extra toppings. Write '0' to finish selecting toppings. \n").strip()
        if pt in pt_dct:
            pizza_topping.append(pt_dct[pt])
        elif pt == '0':
            break                                       
        else:
            print("Sorry, please choose a valid number \n")                

    if pizza_type == "Marinara":
        order = Marinara(Size = pizza_size)
        order.topping = pizza_topping + order.topping
    elif pizza_type == "Margherita":
        order = Margherita(Size = pizza_size)
        order.topping = pizza_topping + order.topping
    elif pizza_type == "TurkPizza":
        order = TurkPizza(Size = pizza_size)
        order.topping = pizza_topping + order.topping
    elif pizza_type == "PlainPizza":
        order = PlainPizza(Size = pizza_size)
        order.topping = pizza_topping + order.topping   

    print(f"Your final total will be {order.get_cost()}.-TL \n")
    order_name = input ("Please enter your name. \n").strip()
    order_ID = input ("Please enter your ID number. \n").strip()
    order_cc = input ("Please enter your credit card number. \n").strip()
    order_ccp = input ("Please enter your credit card password. \n").strip()
    order_time = datetime.datetime.now()
    save_order(Order_db,order,order_name,order_ID,order_cc,order_ccp,order_time)
      
    pizza_topping = []
```

```python
if __name__ == '__main__':
  main()
```
